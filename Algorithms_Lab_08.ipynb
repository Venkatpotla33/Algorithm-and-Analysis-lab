# **Floyd-Warshall_Algorithm**

```python
# Number of vertices
nV = 4
INF = 999
# Algorithm
def floyd(G):
  dist = list(map(lambda p: list(map(lambda q: q, p)), G))
  # Adding vertices individually
  for r in range(nV):
    for p in range(nV):
      for q in range(nV):
        dist[p][q] = min(dist[p][q], dist[p][r] + dist[r][q])
  output_sol(dist)
# Printing the output
def output_sol(dist):
  for p in range(nV):
    for q in range(nV):
      if dist[p][q] == INF:
        print("INF", end=" ")
      else:
        print(dist[p][q], end=" ")
    print(" ") # Move to the next line after printing each row
# Input graph matrix
G = [[0, 5, INF, INF],
     [50, 0, 15, 5],
     [30, INF, 0, 15],
     [15, INF, 5, 0]]
# Run the Floyd-Warshall Algorithm
floyd(G)
```

# **Multi_stage Graph**

```python
N = 12
DIS = float('inf')

graph = [
    [DIS, 9, 7, 3, 2, DIS, DIS, DIS, DIS, DIS, DIS, DIS],
    [DIS, DIS, DIS, DIS, DIS, 4, 2, DIS, DIS, DIS, DIS, DIS],
    [DIS, DIS, DIS, DIS, DIS, 2, 7, DIS, DIS, DIS, DIS, DIS],
    [DIS, DIS, DIS, DIS, DIS, DIS, DIS, 11, DIS, DIS, DIS, DIS],
    [DIS, DIS, DIS, DIS, DIS, DIS, 11, 8, DIS, DIS, DIS, DIS],
    [DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, 6, 5, DIS, DIS],
    [DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, 4, 3, DIS, DIS],
    [DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, 5, 6, DIS],
    [DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, 4],
    [DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, 2],
    [DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, 5],
    [DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS, DIS]
]

def shortest_distance(graph):
    distance = [DIS] * N
    distance[N-1] = 0

    for i in range(N-2, -1, -1):
        for j in range (i+1,N):
            if graph[i][j] == DIS:
                continue
            distance[i] = min(distance[i], graph[i][j] + distance[j])

    return distance[0]

print("Distance from start:", shortest_distance(graph))
```
