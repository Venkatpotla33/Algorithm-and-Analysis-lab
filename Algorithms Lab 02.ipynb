Insertion Sorting


```python
def insertion_sort(arr,n):
  for i in range(1,n):
    # print("iteration",i,"",arr)
    key=arr[i]
    j=i-1
    while j>=0 and arr[j]>key:
      arr[j+1]=arr[j]
      j=1
      arr[j+1]=key
      # print("iteration",i,"",arr)
      return arr

data=[45,32,16,84,76,1,19]
n=len(data)
sorted_data=insertion_sort(data,n)
print("sorted array :",sorted_data)
```

```python
def insertion_sort(arr,n):
  for i in range(1,n):
    print("iteration",i,"",arr)
    key=arr[i]
    j = i - 1  # Initialize j to the last element of the sorted subarray
    while j >= 0 and arr[j] > key:
      arr[j+1]=arr[j]
      j = j - 1 # Decrement j to move to the previous element
    arr[j+1]=key # Insert the key in the correct position
  return arr

data=[45,32,16,84,76,1,19]
n=len(data)
sorted_data=insertion_sort(data,n)
print("sorted array :",sorted_data)
```

Implementing the stack

```python
n=3
top= -1
stack =[None] * n

def push(item):
  global top
  if top >= n-1:
    print("stack is full")

  else:
    top +=1
    stack[top]=item
    print(item,"is pushed into stack")
    print(stack)

def pop():
  global top
  if top == -1:
    return "stack is empty"
  else:
    item=stack[top]
    top -=1
    return item



def display_stack():
  if top == -1:
    print("stack is empty")
  else:
    print("stack elements (top to bottom):")
    for i in range(top,-1,-1):
      print(stack[i])

display_stack()
push(10)
push(20)
push(30)
push(40)
display_stack()
pop()
display_stack()
pop()
pop()
pop()
```
