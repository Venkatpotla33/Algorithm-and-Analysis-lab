```python
def bubble_sort(arr):
    n = len(arr)
    swaps = 0  # Initialize swap counter
    # Traverse through all array elements
    for i in range(n):
        # Last i elements are already in place
        for j in range(0, n - i - 1):
            # traverse the array from 0 to n-i-1
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swaps += 1 # Increment swap counter
    return arr, swaps # Return both the sorted array and the swap count

data = [45, 32, 16, 84, 76, 1, 19]
sorted_data, num_swaps = bubble_sort(data.copy()) # Use a copy to preserve original data and get swap count
print(f"Original data: {data}")
print(f"Sorted data using Bubble Sort: {sorted_data}")
print(f"Number of swaps: {num_swaps}") # Print the number of swaps
```

### Insertion Sort

Insertion Sort is a simple sorting algorithm that builds the final sorted array (or list) one item at a time. It is much less efficient on large lists than more advanced algorithms such as quicksort, heapsort, or merge sort.

Here's the basic idea:

1.  Start with the second element (the first element is considered sorted).
2.  Compare the current element with the elements before it.
3.  If the current element is smaller than the element it's compared to, shift the larger element one position to the right.
4.  Repeat step 3 until the current element is in the correct sorted position or you reach the beginning of the list.
5.  Move to the next unsorted element and repeat the process.

```python
def insertion_sort(arr):
    # Traverse through 1 to len(arr)
    for i in range(1, len(arr)):
        key = arr[i]
        # Move elements of arr[0..i-1], that are
        # greater than key, to one position ahead
        # of their current position
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

data = [45, 32, 16, 84, 76, 1, 19]
sorted_data = insertion_sort(data.copy()) # Use a copy to preserve original data
print(f"Original data: {data}")
print(f"Sorted data using Insertion Sort: {sorted_data}")
```
